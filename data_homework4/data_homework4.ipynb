```python
from collections.abc import Sequence
import random


class QuickSort:
    """
    快速排序类，包含分区和排序功能。
    """

    def __init__(self, arr: Sequence[int | float]):
        """
        初始化快速排序器。

        :param arr: 待排序的数组，可以是整数或浮点数
        """
        self.arr: list[int | float] = list(arr)
        self.length: int = len(arr)

    def partition(self, left_index: int, right_index: int) -> int:
        """
        对数组的指定范围进行分区操作。
        随机选择轴，避免最坏情况

        :param left_index: 分区范围的左边界索引
        :param right_index: 分区范围的右边界索引
        :return: 轴的最终位置索引
        """
        # ========== 修改开始：随机选择轴 ==========
        # 随机选择轴的位置，并与最右元素交换
        pivot_index: int = random.randint(left_index, right_index)
        self.arr[pivot_index], self.arr[right_index] = (
            self.arr[right_index],
            self.arr[pivot_index],
        )
        # ========== 修改结束 ==========

        pivot: int | float = self.arr[right_index]
        right_pointer: int = right_index - 1
        left_pointer: int = left_index

        while True:
            while left_pointer <= right_pointer and self.arr[left_pointer] < pivot:
                left_pointer += 1
            while left_pointer <= right_pointer and self.arr[right_pointer] > pivot:
                right_pointer -= 1
            if left_pointer >= right_pointer:
                break
            self.arr[left_pointer], self.arr[right_pointer] = (
                self.arr[right_pointer],
                self.arr[left_pointer],
            )

        self.arr[left_pointer], self.arr[right_index] = (
            self.arr[right_index],
            self.arr[left_pointer],
        )
        return left_pointer

    def sort(
        self, left_index: int | None = None, right_index: int | None = None
    ) -> list[int | float]:
        """
        对数组进行快速排序（随机轴版本）。
        """
        if left_index is None:
            left_index = 0
        if right_index is None:
            right_index = self.length - 1

        if right_index - left_index <= 0:
            return self.arr

        pivot_position: int = self.partition(left_index, right_index)
        self.sort(left_index, pivot_position - 1)
        self.sort(pivot_position + 1, right_index)

        return self.arr

    # ========== 新增：三路快速排序 ==========
    def sort_3way(
        self, left_index: int | None = None, right_index: int | None = None
    ) -> list[int | float]:
        """
        三路快速排序 - 高效处理重复元素。

        将数组分为三部分：
        - [left_index, lt) : 小于轴的元素
        - [lt, gt]         : 等于轴的元素
        - (gt, right_index]: 大于轴的元素

        :param left_index: 排序范围的左边界索引
        :param right_index: 排序范围的右边界索引
        :return: 排序后的数组
        """
        if left_index is None:
            left_index = 0
        if right_index is None:
            right_index = self.length - 1

        if right_index <= left_index:
            return self.arr

        # 随机选择轴并交换到最左边
        pivot_index: int = random.randint(left_index, right_index)
        self.arr[left_index], self.arr[pivot_index] = (
            self.arr[pivot_index],
            self.arr[left_index],
        )
        pivot: int | float = self.arr[left_index]

        # 三路分区指针
        lt: int = left_index      # arr[left_index..lt-1] < pivot
        gt: int = right_index     # arr[gt+1..right_index] > pivot
        i: int = left_index + 1   # 当前扫描位置

        while i <= gt:
            if self.arr[i] < pivot:
                # 小于pivot，放到左边
                self.arr[lt], self.arr[i] = self.arr[i], self.arr[lt]
                lt += 1
                i += 1
            elif self.arr[i] > pivot:
                # 大于pivot，放到右边（i不增加，需检查交换过来的元素）
                self.arr[i], self.arr[gt] = self.arr[gt], self.arr[i]
                gt -= 1
            else:
                # 等于pivot，留在中间
                i += 1

        # 递归排序小于和大于的部分（等于部分已就位）
        self.sort_3way(left_index, lt - 1)
        self.sort_3way(gt + 1, right_index)

        return self.arr
```
