```python
class HashDict:
    def __init__(self,size: int):
        """
        :param size: 列表长度
        """
        self.size: int = size
        self.table: list[list[tuple[str, int]]] = [[] for _ in range(size)]

    def put(self,key: str,value: int)->None:
        """
        向字典中增加元素
        :param key: 键
        :param value: 值
        """
        hash_result:int =hash(key) % self.size
        bucket:list[tuple[str,int]] = self.table[hash_result]

        for index,(old_key,old_value) in enumerate(bucket):
            #如果有旧值就替换
            if old_key == key:
                bucket[index] = (key,value)
                return
        #如果所有位置没有旧值，就追加
        else:
            bucket.append((key,value))

    def get(self,key:str):
        """
        获取键对应的值
        :param key: 键
        :return: 值，如果不存在返回None
        """
        hash_result:int =hash(key) % self.size
        bucket:list[tuple[str,int]] = self.table[hash_result]
        for bucket_key,bucket_value in bucket:
            if bucket_key == key:
                return bucket_value
        return None
    def delete(self,key:str)->None:
        """
        删除键对应的值
        :param key: 键
        """
        hash_result:int =hash(key) % self.size
        bucket:list[tuple[str,int]] = self.table[hash_result]
        for index,(bucket_key,bucket_value) in enumerate(bucket):
            if bucket_key == key:
                del bucket[index]

        return None

    def __str__(self)->str:
        """
        打印字典
        :return: 字典的字符串表示
        """
        items={}
        for bucket in self.table:
            for key,value in bucket:
                items[key]=value
        return str(items)
```

```python
hash_dict=HashDict(size=100000)
hash_dict.put('abc',1001)
hash_dict.put('acd',1002)
hash_dict.put('abd',1003)
hash_dict.get('acd')
```
